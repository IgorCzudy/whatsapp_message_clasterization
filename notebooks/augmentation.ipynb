```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../CleanChats/FullChat.csv")
```

```python
df.iloc[8:30, ]
```

```python
from nlpaug.augmenter.char import KeyboardAug
```

```python
input = list(df.head()['Message'])
new_messages = KeyboardAug( aug_char_max=1, aug_word_max=2, include_upper_case=False, include_special_char=False, include_numeric=False).augment(input)
pd.DataFrame({"Message": new_messages, "Section": df.head()["Section"]})
```

```python
from nlpaug.augmenter.word import SynonymAug

SynonymAug(aug_src='wordnet').augment(input)
```

```python
from nlpaug.augmenter.word import ReservedAug
clubs_list = [["La3 Club", "Chocolate", "Espit Chupitos", "Barraca", "Santoory", "Umbracle", "Bamboo Pub", "Mya Disco", "House Music Collective", "Club Gordo", "Indiana Disco", "Natura Dub"]]

ReservedAug(reserved_tokens=clubs_list).augment(df.iloc[8, :]["Message"], n=50)
```
